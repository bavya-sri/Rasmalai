```python
import imageio as iio
import matplotlib.pyplot as plt
import skimage.filters
import numpy as np
import cv2
%matplotlib widget
```

```python
image = iio.imread(uri="car.jpg")

# display the image
fig, ax = plt.subplots()
plt.imshow(image)
```

```python
sigma = 3.0

# apply Gaussian blur, creating a new image
blurred = skimage.filters.gaussian(
    image, sigma=(sigma, sigma), truncate=3.5)

# display blurred image
fig, ax = plt.subplots()
plt.imshow(blurred)
```

```python
picture = cv2.imread("car.jpg")
picture = cv2.cvtColor(picture, cv2.COLOR_BGR2RGB)
addingNoise = np.zeros(picture.shape, np.uint8)
```

```python
output = [image, blurred]

titles = ['Original Image', 'Gaussian blur Image']

for i in range(2):
    plt.subplot(1, 2, i + 1)
    plt.imshow(output[i])
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])
plt.show()
```
