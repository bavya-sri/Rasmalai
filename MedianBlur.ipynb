```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
imagePath = '/content/drive/MyDrive/DataScience/train/106025.jpg'
picture = cv2.imread(imagePath)
picture = cv2.cvtColor(picture, cv2.COLOR_BGR2RGB)
```

```python
addingNoise = np.zeros(picture.shape, np.uint8)
```

```python
val = 0.2

for i in range(picture.shape[0]): 
    for j in range(picture.shape[1]): 
        r = random.random()
        if r < val / 2:
            addingNoise[i][j] = [0, 0, 0] #black noise
        elif r < val:
            addingNoise[i][j] = [255, 255, 255] #white noise
        else:
            addingNoise[i][j] = picture[i][j] #original image pixel
```

```python
removeNoise = cv2.medianBlur(addingNoise, 5)
```

```python
output = [picture, addingNoise, removeNoise]

titles = ['Original Image', 'Noisy Image', 'Denoised Image']

for i in range(3):
    plt.subplot(1, 3, i + 1)
    plt.imshow(output[i])
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])
plt.show()
```
