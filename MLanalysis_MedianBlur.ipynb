```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
imagePath = r'C:\Users\sgudur3\Desktop\418 IDS\Traffic signals\istockphoto-1057895398-612x612.jpg'
picture = cv2.imread(imagePath)
picture = cv2.cvtColor(picture, cv2.COLOR_BGR2RGB)
```

```python
addingNoise = np.zeros(picture.shape, np.uint8)
```

```python
val = 0.2

#traversing throughout the image pixels
for i in range(picture.shape[0]): #rows
    for j in range(picture.shape[1]): #cols
        r = random.random()
        if r < val / 2:
            addingNoise[i][j] = [0, 0, 0] #black noise
        elif r < val:
            addingNoise[i][j] = [255, 255, 255] #white noise
        else:
            addingNoise[i][j] = picture[i][j] #original image pixel
```

```python
# median blur
removeNoise = cv2.medianBlur(addingNoise, 5)
```

```python
output = [picture, addingNoise, removeNoise]

titles = ['Original Image', 'Noisy Image', 'Denoised Image']

for i in range(3):
    plt.subplot(1, 3, i + 1)
    plt.imshow(output[i])
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])
plt.show()
```
